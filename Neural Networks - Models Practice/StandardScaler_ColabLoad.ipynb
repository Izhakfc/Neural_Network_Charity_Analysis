```python
# Import our dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Read in our dataset
path = "/content/gdrive/MyDrive/Neural Networks/Resources"
hr_df = pd.read_csv(f'{path}/hr_dataset.csv')
hr_df.head()
```

```python
# Create the StandardScaler instance -> Fit -> Transform
scaler = StandardScaler()
```

```python
# Fit the StandardScaler
scaler.fit(hr_df)
```

```python
# Scale the data
scaled_data = scaler.transform(hr_df)
```

```python
# Create a DataFrame with the scaled data // Scaler returns an array
transformed_scaled_data = pd.DataFrame(scaled_data, columns=hr_df.columns)
transformed_scaled_data.head()
```
